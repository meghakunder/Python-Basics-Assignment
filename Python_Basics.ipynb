**Question 2 : Explain the difference between List and Tuple in Python.**

**ANS:**



In Python, both **lists** and **tuples** are data structures used to store collections of items. They can hold elements of different data types such as integers, strings, and even other collections. However, they differ in several important aspects.

 1. Mutability

The primary difference between a list and a tuple is **mutability**.

* A **list is mutable**, meaning its elements can be modified after creation.
* A **tuple is immutable**, meaning once it is created, its elements cannot be changed.

**Example:**

my_list = [1, 2, 3]

my_list[0] = 10   # Allowed

my_tuple = (1, 2, 3)

my_tuple[0] = 10  # Error

 2. Syntax

Lists and tuples are defined using different symbols:

* Lists use **square brackets []**
* Tuples use **parentheses ()**

**Example:**


list_example = [1, 2, 3]

tuple_example = (1, 2, 3)

 3. Performance

Tuples are generally **faster than lists** because they are immutable. Since their values cannot change, Python can optimize their performance.

 4. Memory Usage

Tuples consume **less memory** compared to lists. This makes them more efficient when working with large amounts of fixed data.

 5. Use Cases

* **Lists** are used when data needs to be modified frequently (e.g., adding, removing, or updating elements).
* **Tuples** are used when data should remain constant (e.g., coordinates, fixed configurations).

 6. Built-in Methods

* Lists have many built-in methods like append(), remove(), sort(), etc.
* Tuples have fewer methods since they cannot be modified.

**Question 8 : Write a Python program to remove duplicate values from a list using a set**
